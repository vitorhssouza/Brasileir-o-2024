```python
import requests
import json
import pandas as pd
from pyspark.sql.functions import col
import time
```

```python
# Criando um dataframe no spark

from pyspark.sql import SparkSession

# Iniciando uma SparkSession
spark = SparkSession.builder.appName("dm_jogadores").getOrCreate()
```

```python
# Lendo a dm_clube e transformando em um df
df = pd.read_parquet('/lakehouse/default/Files/Bronze/dm_clubes')
```

```python
jogadores_clubes = list()
joga = dict()


for c in (df['id']):
    uri = f'https://api.football-data.org/v4/teams/{c}'
    headers = { 'X-Auth-Token': 'YOUR_API_TOKEN' }
    response = requests.get(uri, headers=headers)

    for jog in range(0, len(response.json()['squad'])):
        joga = {
              'id_clube': response.json()['id'],
              'id_jogadores': response.json()['squad'][jog]['id'],
              'nome_jogadores': response.json()['squad'][jog]['name'],
              'posicao': response.json()['squad'][jog]['position'],
               'nacionalidade': response.json()['squad'][jog]['nationality']
     }
        jogadores_clubes.append(joga)
    time.sleep(10)

```

```python
# Criando dataframe 
df_jogadores = spark.createDataFrame(jogadores_clubes)
```

```python
# Salvando o dataframe em um arquivo parquet
df_jogadores.write.mode('overwrite').parquet('Files/Bronze/dm_jogadores')
```
