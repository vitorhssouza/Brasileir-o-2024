### Criando tabela de partidas

```python
# importando bibliotecas

import requests
import json
import pandas as pd

```

```python
# Criando um dataframe no spark

from pyspark.sql import SparkSession

# Iniciando uma SparkSession
spark = SparkSession.builder.appName("dm_partida").getOrCreate()

uri = 'https://api.football-data.org/v4/competitions/2013/matches'
headers = { 'X-Auth-Token': 'YOUR_API_TOKEN' }

response = requests.get(uri, headers=headers)
```

```python
partidas = list()
parti = dict()
for c in range(0, len(response.json()['matches'])): 
    parti  = {
        'Status': response.json()['matches'][c]['status'],
        'Data'        : response.json()['matches'][c]['season']['startDate'],
        'Rodada': response.json()['matches'][c]['matchday'],
        'Time Casa': response.json()['matches'][c]['homeTeam']['name'],
        'Resultado': f"{response.json()['matches'][c]['score']['fullTime']['home']} x {response.json()['matches'][c]['score']['fullTime']['away']}",
        'Time Visitante': response.json()['matches'][c]['awayTeam']['name']


    }
    partidas.append(parti)
```

```python
# Criando um dataframe pandas pyspark
df = spark.createDataFrame(partidas)

#reordenando as colunas de acordo com a lista
df = df.select("Status", "Data", "Rodada", "Time Casa", "Resultado", "Time Visitante")
#df.show()
```

```python
# Salvando o DataFrame em um arquivo parquet
df.write.mode("overwrite").parquet('Files/Bronze/dm_partidas')
```
