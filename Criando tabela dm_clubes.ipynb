### Noteebok para criação de tabelas com os id dos clube na api 

```python
import requests
import json
import pandas as pd
```

```python
# Criando um dataframe no spark

from pyspark.sql import SparkSession

# Iniciando uma SparkSession
spark = SparkSession.builder.appName("dm_clubes").getOrCreate()
```

```python
uri = 'https://api.football-data.org/v4/competitions/BSA/standings'
headers = { 'X-Auth-Token': 'YOUR_API_TOKEN' }

response = requests.get(uri, headers=headers)
```

```python
dm_clubes = list()
nome_clube = dict()


for c in range(len(response.json()['standings'][0]['table'])): 
    
    nome_clube = {
        'id': response.json()['standings'][0]['table'][c]['team']['id'],
        'nome': response.json()['standings'][0]['table'][c]['team']['name']
    }

    dm_clubes.append(nome_clube)
```

```python
df = spark.createDataFrame(dm_clubes)
```

```python
# Salvando o DataFrame em um arquivo parquet

df.write.mode("overwrite").parquet('Files/Bronze/dm_clubes')
```
