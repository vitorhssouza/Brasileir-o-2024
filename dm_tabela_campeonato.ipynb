### Criando tabela dimensão da tabela do campeonato brasileiro 2024

```python
# importando bibliotecas

import requests
import json
import pandas as pd
```

```python
# Criando um dataframe no spark

from pyspark.sql import SparkSession

# Iniciando uma SparkSession
spark = SparkSession.builder.appName("dm_tabela_campeonato").getOrCreate()

uri = 'https://api.football-data.org/v4/competitions/BSA/standings'
headers = { 'X-Auth-Token': '5YOUR_API_TOKEN' }

response = requests.get(uri, headers=headers)
```

```python
#response.json()
```

```python
#response.json()['standings'][0]['table'][0]

posicao = list()
posicao_dict = dict()

for p in range(len(response.json()['standings'][0]['table'])):
    posicao_dict = {
        'Id_clube': response.json()['standings'][0]['table'][p]['team']['id'],
        'Posição' : response.json()['standings'][0]['table'][p]['position'],
        'Clube'   : response.json()['standings'][0]['table'][p]['team']['name'],
        'Pontos'  : response.json()['standings'][0]['table'][p]['points'],
        'Total_de_jogos': response.json()['standings'][0]['table'][p]['playedGames'],
        'Vitoria': response.json()['standings'][0]['table'][p]['won'],
        'Empate' : response.json()['standings'][0]['table'][p]['draw'],
        'Derrota': response.json()['standings'][0]['table'][p]['lost'],
        'GP': response.json()['standings'][0]['table'][p]['goalsFor'],
        'GC': response.json()['standings'][0]['table'][p]['goalsAgainst'],
        'Saldo_gols': response.json()['standings'][0]['table'][p]['goalDifference']
    }
    posicao.append(posicao_dict)
    #print(response.json()['standings'][0]['table'][p])
```

'won': 12,

 'points': 39,
 'goalsFor': 29,
 'goalsAgainst': 14,
 'goalDifference': 15}

```python
# Criando um dataframe pandas pyspark
df = spark.createDataFrame(posicao)
#df.show()
```

```python

# Criando um arquivo parquet
df.write.mode("overwrite").parquet('Files/Bronze/dm_tabela')
```
