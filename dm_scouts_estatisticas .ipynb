#### Análise dos artilheiros do brasileirão 2024

```python
import requests
import json
import pandas as pd

# Criando um dataframe no spark

from pyspark.sql import SparkSession

# Iniciando uma SparkSession
spark = SparkSession.builder.appName("dm_artilheiro").getOrCreate()

uri = 'https://api.football-data.org/v4/competitions/2013/scorers'
headers = { 'X-Auth-Token': 'YOUR_API_TOKEN' }

response = requests.get(uri, headers=headers)

```

```python
artilheiro = list()
art = dict()

for c in range(0, len(response.json()['scorers'])):
    art = {
        'Id_jogador': response.json()['scorers'][c]['player']['id'],
        'Position': c+1,
        'Name':     response.json()['scorers'][c]['player']['name'],
        'Team':     response.json()['scorers'][c]['team']['id'],
        'PlayedMatches': response.json()['scorers'][c]['playedMatches'],
        'Goals': response.json()['scorers'][c]['goals'],
        'Assists': response.json()['scorers'][c]['assists'], 
        'Penalties': response.json()['scorers'][c]['penalties']
    }
    artilheiro.append(art)
```

```python

# Criando um dataframe pandas pyspark
df = spark.createDataFrame(artilheiro)
#df.show()

# Reajustando coluna
df = df.select("Id_jogador", "Position", "Name", "Team", "PlayedMatches", "Goals", "Assists", "Penalties")

# Salvando o DataFrame em um arquivo parquet

df.write.mode("overwrite").parquet('Files/Bronze/dm_artilheiros')
```
